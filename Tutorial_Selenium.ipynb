```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
#Standaardcode voor het vinden de chrome webdriver. Dit is NIET hetzelfde als de Chrome Browser
```

```python
help(By)
```

```python
driver = webdriver.Chrome(PATH) #Start de driver op.
driver.get("https://en.wikipedia.org/wiki/Main_Page") #Laad de driver navigeren naar een gegeven URL
search = driver.find_element("name", "search") #Geeft slechts één element terug als individuele waarde, niet als lijst
search.send_keys(["James Bond", Keys.RETURN])
```

```python
driver.title
```

```python
#Stap 1: Zoek naar een "moederelement", waarvan je zeker weet dat als die geladen is, elk onderliggend relevant child-element
#ook geladen is.

#Stap 2: Laat het "moederelement in en wijs deze toe aan een variabele". In dit geval wordt er gezocht naar een ID.

try:
    all_search_results = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'mw-content-text'))
    )
except:
    print("Search results were not found")
```

```python
print(all_search_results.text)
```

```python
all_search_result_titles = all_search_results.find_elements(By.CLASS_NAME, "mw-search-result-heading")
#In tegenstelling tot By.NAME van net geeft By.CLASS_NAME een hele lijst terug.

for search_result_title in all_search_result_titles:
    print(search_result_title.text)
```

```python
link = driver.find_element(By.LINK_TEXT , "James Bond")
link.click()
#.click() werkt ook voor buttons die je bijvoorbeeld vindt met
#By.ID, By.NAME & By.CLASS_NAME
```

```python
try:
    all_page_content = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'mw-page-container'))
    )
except:
    print("Page was not found")

all_page_content.text
```
